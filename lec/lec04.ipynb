- 기존에 1차원 일때

$$ H(x) = Wx + b $$
    
- 다차원 Hypothesis
$$ H(x_1, x_2, x_3) = w_1x_1 + w_2x_2 + w_3x_3 + b $$

- Cost Function
$$ cost(W,b) = \frac{1}{m}\sum_{i=1}^m(H(x_1^i,x_2^i,x_3^i)-y^i)^2 $$

- 더 많으면?
$$ H(x_1, x_2, x_3,...,x_n) = w_1x_1 + w_2x_2 + w_3x_3 + ... + w_nx_n + b $$
    
> 겁나 많아서 불편함!! 

- Matrix

\begin{equation*}
\begin{pmatrix}x_1,x_2,x_3\end{pmatrix} 
* 
\begin{pmatrix}w_1\\w_2\\x_3\end{pmatrix}
=
\begin{pmatrix}x_1w_1,x_2w_2,x_3w_3\end{pmatrix} 
\end{equation*}
    
$$ H(X) = XW $$
  
- 인스턴스가 많다면?
\begin{equation*}
\begin{pmatrix}
x_{11},x_{12},x_{13}\\

x_{51},x_{52},x_{53}

\end{equation*}
