- Cost Function

> H(X)가 sigmoid function이면, cost function의 모양은 sigmoid가 여러 개인 형태를 띈다. 즉, 경사하강 방식으로 최솟값을 찾는다면, global minimum(전체에서의 최솟값) 값이 아닐 수 있다.

- New cost function for logistic

$$ cost(W) = \frac{1}{m}\sum_{} c(H(x),y) $$

- x 범위에 따라 나누어서 표현하는 방법을 모르게씀...ㄷㄷ
$$ c(H(x),y) = -log(H(x)), y=1 $$
$$ -log(H(1-x)), y=0 $$

```python
import matplotlib.pyplot as plt
import numpy as np

# zerox 는 y=0
zerox = np.arange(-10, 10, 0.01)

# onex 는 y=1
onex = np.arange(-10, 10, 0.01)

zerohp = 1/(1+np.exp(-zerox))
onehp = 1/(1+np.exp(-onex))

# y=0일 때, 가설이 0이면 cost가 0, 1이면 cost 무한대
zerocost = -np.log(1-zerohp)
plt.title("if y=0")
plt.xlabel("prediction")
plt.ylabel("cost")
plt.plot(zerohp, zerocost)
plt.show()

# y=1일 때, 가설이 1이면 cost가 0, 0이면 cost 무한대
onecost = -np.log(onehp)
plt.title("if y=1")
plt.xlabel("prediction")
plt.ylabel("cost")
plt.plot(onehp, onecost)
plt.show()
```

- if condition을 없애보자.

$$ C:(H(x),y) = -ylog(H(x))-(1-y)log(1-H(x)) $$

$$ y=1, C = -log(H(x)) $$
$$ y=0, C = -log(1-H(x)) $$

- Cost를 Minimize하자

$$ cost(W) = -\frac{1}{m}\sum_{}ylog(H(x))+(1-y)log(1-H(x))$$

> 경사 하강!!

$$ W := W - \alpha\frac{\delta}{{\delta}W}cost(W) $$
